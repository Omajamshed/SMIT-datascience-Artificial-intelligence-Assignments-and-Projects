```python
# Take input from the user for customer name, age, account number, and initial deposit.
customername=input("enter your full name")
age=int(input("enter your age:"))
AN=input("enter your account number:")
initial_deposit = float(input("Enter your initial deposit amount: "))
# Display the data type of each input using.
print(type(customername))
print(type(age))
print(type(AN))
print(type(initial_deposit))
```

#  Perform string operations:


```python
#  Print the customer name in uppercase
#  Count and display the number of characters in the name
print(customername.upper())
print(len(customername))
```

# Apply type casting and conversion:


```python
#  Convert account number to string and display its length
#  Convert age into float and display it
AN_str = str(AN)
print("Account Number as String:", AN_str)
print("Length of Account Number:", len(AN_str))
```

# Use data structures:


```python
#  A list to store the last three deposit/withdrawal amounts
#  A tuple to store available account types (e.g., Savings, Current, Fixed)
#  A dictionary to store customer details (name, age, account number, balance)
transactions = [5000, -2000, 3000] 
print("Last 3 Transactions:", transactions)
# Tuple for available account types
available_accounts = ("Savings", "Current", "Fixed")
print("Available Account Types:", available_accounts)

# Dictionary 
customer = {
    "Name": customername,
    "Age": age,
    "Account Number": AN_str,
    "Balance": initial_deposit
}
print("Customer Details:", customer)
```

# Perform mathematical operations:


```python
#  Add deposits to balance
#  Subtract withdrawals from balance
#  Calculate interest using a formula (e.g., balance × 0.05)
#  Use modulus operator to check if balance is an even or odd number
balance = initial_deposit + transactions[0] + transactions[2] - abs(transactions[1])
print("Updated Balance after Transactions:", balance)
# Calculate interest
interest = balance * 0.05
print("Interest (5%):", interest)
# Check even or odd balance
if int(balance) % 2 == 0:
    print("Balance is Even")
else:
    print("Balance is Odd")
```

# Use conditional structures to determine:


```python
#  If balance ≥ 100,000 → print “Premium Account Holder”
#  If balance ≥ 50,000 → print “Gold Account Holder”
#  If balance ≥ 20,000 → print “Silver Account Holder”
#  Otherwise → print “Basic Account Holder”
balance=int(input("enter your baalance:"))
if balance >= 100000:
    print("Premium Account Holder")
elif balance >= 50000:
    print("Gold Account Holder")
elif balance >= 20000:
    print("Silver Account Holder")
else:
    print("Basic Account Holder")
```


# Use while loops:


```python

#  Keep asking for the correct bank login password until entered correctly
correct_password = "1234"
entered_password = ""
while entered_password != correct_password:
    entered_password = input("Enter bank password to continue: ")
    if entered_password != correct_password:
        print("Incorrect password! Try again.")
print("Login Successful!")
```

```python
#  Print numbers from 1 to 10 as a simple “transaction counter” at the end
counter = 1
while counter <= 10:
    print("Transaction", counter)
    counter += 1

print("\nThank you for using our Bank Account Management System!")
```
